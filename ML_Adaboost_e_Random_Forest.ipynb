```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder

from google.colab import files
uploaded = files.upload()
```

### Carga dos dados e particionamento das bases de treinamento e teste

### Transformação de dados

A classe é convertida para labels únicos sequenciais.

<code>
 le = preprocessing.LabelEncoder()
  
 le.fit(dados)
</code>


### Particionamento da base

<code>train_test_split(X, y) -- particiona a base de dados original em bases de treinamento e teste.</code>

No código a seguir, são utilizados 10% para teste e 90% para treinamento.







```
sonar = pd.read_excel('sonar.xlsx', sheet_name=0) 
print("\nDimensões: {0}".format(sonar.shape))
print("\nCampos: {0}".format(sonar.keys()))
print(sonar.describe(), sep='\n')

X = sonar.iloc[:,0:(sonar.shape[1] - 1)]

le = LabelEncoder()
y = le.fit_transform(sonar.iloc[:,(sonar.shape[1] - 1)])

class_names = le.classes_
print("\nClasses: {0}".format(class_names))

# Particiona a base de dados
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)


```

## Indução do Modelo 


### Árvore de Decisão

```
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Classificador Árvore de Decisão:\n")
print(classification_report(y_test, y_pred, target_names=class_names))

```

### Random forest com 10 Árvores

```
clr = RandomForestClassifier(n_estimators=31, random_state=10)
clr = clf.fit(X_train, y_train)
y_pred = clr.predict(X_test)
print("Classificador Random Forest:\n RandomForestClassifier(n_estimators=31)\n")

print(classification_report(y_test, y_pred, target_names=class_names))
```

```
list_of_score = []
for n in range (2, 40, 2):
  clr = RandomForestClassifier(n_estimators=n, random_state=10)
  clr = clf.fit(X_train, y_train)
  y_pred = clr.predict(X_test)
  list_of_score.append({'n': n, 'score': clr.score(X_test, y_test)})

print(list(reversed(sorted(list_of_score, key=lambda k: k['score']))))
```

### Random forest com heurísticas extremas

Varia o tamanho das sub-amostras.


```
cle = ExtraTreesClassifier(n_estimators=5, random_state=0)
cle = cle.fit(X_train, y_train)
y_pred = cle.predict(X_test)
print("Classificador Extreme Tree:\n ExtraTreesClassifier(n_estimators=5)\n")
print(classification_report(y_test, y_pred, target_names=class_names))
```

### Adaboost com árvores mínimas


```
ada = AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=5), 
                         algorithm="SAMME", 
                         n_estimators=50)
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
print("Classificador AdaBoost:\n AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=1), algorithm=\"SAMME\", n_estimators=50)\n")
print(classification_report(y_test, y_pred, target_names=class_names))
```

```
ada = AdaBoostClassifier(RandomForestClassifier(n_estimators=31, random_state=10), 
                         algorithm="SAMME", 
                         n_estimators=50)
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
print("Classificador AdaBoost:\n AdaBoostClassifier(RandomForestClassifier, algorithm=\"SAMME\", n_estimators=5)\n")
print(classification_report(y_test, y_pred, target_names=class_names))
```

```
from sklearn.svm import SVC
ada = AdaBoostClassifier(SVC(gamma='scale'), 
                         algorithm="SAMME", 
                         n_estimators=50)
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)
print("Classificador AdaBoost:\n AdaBoostClassifier(SVC(gamma='scale'), algorithm=\"SAMME\", n_estimators=5)\n")
print(classification_report(y_test, y_pred, target_names=class_names))
```

### SVM

```
from sklearn.svm import SVC
svm_sonar = SVC(gamma='scale')
svm_sonar = svm_sonar.fit(X_train, y_train)
y_pred = svm_sonar.predict(X_test)
print(classification_report(y_test, y_pred, target_names=class_names))
```
