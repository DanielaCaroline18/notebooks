## Regras de associação geradas a partir de itemsets frequentes



```

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

from google.colab import files
uploaded = files.upload()
```

**Carregando base de dados**

```
dataset = pd.read_excel('Comunidade.xlsx', sheet_name=0) 
print("\nDimensões: {0}".format(dataset.shape))
print("\nCampos: {0}".format(dataset.keys()))
print(dataset.describe(), sep='\n')
```

**Separando apenas as colunas que serão utilizadas para gerar o modelo**

```
datasetModificado= dataset.iloc[:, 5:12]
print("\nDimensões: {0}".format(datasetModificado.shape))
print("\nCampos: {0}".format(datasetModificado.keys()))
print(datasetModificado.describe(), sep='\n')
#print(datasetModificado)
```

**Transformando a base em DataFrame**

Como a base modificada obtida já era booleana (binomial), basta transformá-la em um Data Frame
para que possa ser manipulada

```
columns = ['Familiar', 'Hobbies', 'Clubes', 'Política', 'Profissional','Religiosa', 'Grupo de suporte']

df = pd.DataFrame(datasetModificado, columns=columns)
```

```
#print("Base transformada (binomial):\n{}\n".format(df))  
#print("Data frame:\n{}\n".format(df))
```

## **Utilizando Apriori**

```


frequent_itemsets_apriori = apriori(df, min_support=0.2, use_colnames=True)

print(frequent_itemsets_apriori)
```

## **Utilizando fpgrowph**

```
frequent_itemsets_fpgrowth  = fpgrowth(df, min_support=0.1, use_colnames=True)
print(frequent_itemsets_fpgrowth)
```

Ao diminuir o suporte para 0.1 o número de itemsets frequentes aumenta para 16, enquanto que para 0.2 o número de itemsets é apenas 5. Por se tratar de uma base relativamente pequena o aumento no número de itemsets pode-se tratar apenas de ruídos, mas para se avaliar as regras foi feita a escolha de usar o melhor suporte.

```
association_rules(frequent_itemsets_fpgrowth, metric="confidence", min_threshold=0.6)
```

Gera regras de associação com lift mínimo de 1.3 


```
#Linha adicionada somente para gerar uma saida mais bem formatada (linha desnecessária)
association_rules(frequent_itemsets_fpgrowth, metric="lift", min_threshold=2)
```

Pelas regras geradas percebe-se que a relação "Familiar" e "Religiosa" estão bastante ligadas, pois existem muitas associações que possuem essa relação. "Hobbie" e "Clubes" também estão associados e possuem um alto valor de confiança.
